# Compton Scattering

## Theory

A high frequency photon can be scattered if it interacts with a electronic state that is occupied. Compton derived the mathematical relationship between the change in wavelength and the scattering angle of a photon by a stationary electron:

$$\Delta \lambda = \frac{h}{m_e c}(1-\cos (\theta))$$

- $\lambda$ is the wavelength of the photon
- $h$ is Planck's constant
- $m_e$ is the mass of the electron
- $c$ is the speed of light
- $\theta$ is the scattering angle of the photon

If the electron isn't stationary then there will be doppler broadening, related to the momentum of the electron. A given momentum state can only scatter if its occupied. Encoded in this distribution is information about the occupied states meaning the fermi surface can be measured. By measuring the energy of scattered photons under different magnetization two different compton profiles can be generated ibe fir soub up and one for spin down. 

$$J(p_z) = \int \int \rho (p_x, p_y, p_z) dp_x dp_y$$

Strengths:
- Provides direct visualization of the 3D fermi surface
- Electron mean-free path independent

Weaknesses:
- Requires high energy photons (typically from a synchrotron source)
- **Data analysis can be complex due to the need to deconvolute instrumental effects and account for multiple scattering events.**
- Slow measurments since lots of measurments need to be carried out to go from 1D to 3D.
