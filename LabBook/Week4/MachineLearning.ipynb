## Theory

Assume the noisy image can be written as a matrix $P_{i}$ where each element represents the pixel intensity of the $i$ th pixel. The goal is to train a neural network $f(P; K)$ where $K$ is the known convolution kernel to output a denoised image $Q_{i} = f(P; K)$.



The node $a^{(1)}_0$ will have its value computed from a sum of the weighted inputs from the first layer.
$$a^{(1)}_0 = \sigma(b_0+\sum_{i} w_{i}P_{i})$$
Where $\sigma$ is the normalisation function given by $\sigma(x) = \frac{1}{1 + e^{-x}}$. $w_{i}$ are the weights associated with each input pixel $P_{i}$, $b_0$ is the bias term.

The first layers of hidden nodes $a^{(1)}$ can be written as $$a^{(i+1)} = \sigma(W^{(i)}a^{(i)} + b^{(i)})$$ where $W^{(i)}$ is the weight matrix and $b^{(i)}$ is the bias vector which transforms the $i$ th layer into the $i+1$ th layer. The values of $W^{(i)}$ and $b^{(i)}$ are learned during training.
